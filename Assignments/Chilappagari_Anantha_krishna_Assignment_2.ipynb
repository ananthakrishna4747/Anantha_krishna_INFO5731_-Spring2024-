# Question 1 (40 points)

Write a python program to collect text data from **either of the following sources** and save the data into a **csv file:**

(1) Collect all the customer reviews of a product (you can choose any porduct) on amazon. [atleast 1000 reviews]

(2) Collect the top 1000 User Reviews of a movie recently in 2023 or 2024 (you can choose any movie) from IMDB. [If one movie doesn't have sufficient reviews, collect reviews of atleast 2 or 3 movies]

(3) Collect all the reviews of the top 1000 most popular software from G2 or Capterra.

(4) Collect the **abstracts** of the top 10000 research papers by using the query "machine learning", "data science", "artifical intelligence", or "information extraction" from Semantic Scholar.

(5) Collect all the information of the 904 narrators in the Densho Digital Repository.


```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

page_range = range(1, 41)
base_url = "https://ddr.densho.org/narrators/?page="
headers = {
    'User-Agent':
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
urls = [base_url + str(page_num) for page_num in page_range]

data = []

for url in urls:
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    a_tags = soup.find_all('a', class_='item-hover')

    for a_tag in a_tags:
        url2 = a_tag['href']
        response2 = requests.get(url2, headers=headers)
        soup2 = BeautifulSoup(response2.content, "html.parser")

        try:
            h1_tag = soup2.find('h1').text.strip() if soup2.find('h1') else ''
            p_tag = soup2.find('p').text.strip() if soup2.find('p') else ''
            h4_tag = soup2.find('h4').text.strip() if soup2.find('h4') else ''
            source_tag = soup2.find('div', class_='source muted').text.strip() if soup2.find('div', class_='source muted') else ''
            source_lines = [line.strip() for line in source_tag.split('\n') if line.strip()]
            img_tag = soup2.find('img', class_='img-responsive')['src'] if soup2.find('img', class_='img-responsive') else ''
            inter_class = soup2.find('a', class_='pull-left')['href']
            response3 = requests.get(inter_class, headers=headers)
            soup3 = BeautifulSoup(response3.content, "html.parser")
            vid_tags = soup3.find_all('a', class_='btn btn-default btn-xs')

            data.append({
                'URL': a_tag['href'],
                'Name': h1_tag,
                'Description': p_tag,
                'Personal_page': inter_class,
                'Image_URL': img_tag,
                'Interview_Date': source_lines[0] if len(source_lines) > 0 else '',
                'Interview_Location': source_lines[1] if len(source_lines) > 1 else '',
                'Interview_Length': source_lines[2].replace('—', '').strip() if len(source_lines) > 2 else '',
                'No_of_Segments': source_lines[3].replace('segments.', 'vs').strip() if len(source_lines) > 3 else '',
                'Interview_Link': vid_tags[2]['href'] if len(vid_tags) > 2 else ''
            })

        except Exception as e:
            print("Error processing page:", url2)
            print("Error:", e)
            print("-" * 50)

df = pd.DataFrame(data)
df.to_csv('output.csv', index=True)
print("Data saved to 'output.csv'")
df = pd.read_csv('/content/output.csv')
df
```

# Question 2 (30 points)

Write a python program to **clean the text data** you collected in the previous question and save the clean data in a new column in the csv file. The data cleaning steps include: [Code and output is required for each part]

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the stopwords list.

(4) Lowercase all texts

(5) Stemming.

(6) Lemmatization.

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from nltk.stem import PorterStemmer, WordNetLemmatizer
import string

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
def remove_special_characters(text):
    if isinstance(text, (int, float)):  # Check if input is an integer or float
        return str(text)  # Convert to string and return
    return ''.join([char for char in text if char.isalnum() or char.isspace()])

def remove_numbers(text):
    if isinstance(text, (int, float)):  # Check input
        return str(text)  # Convert to string and return
    return ''.join([char for char in text if not char.isdigit()])

def remove_stopwords(text):
    if isinstance(text, (int, float)):  # Check input
        return str(text)  # Convert to string and return
    return ' '.join([word for word in text.split() if word.lower() not in stopwords.words('english')])

def lowercase_text(text):
    if isinstance(text, (int, float)):  # Check input
        return str(text)  # Convert to string and return
    return text.lower()

def stem_text(text):
    if isinstance(text, (int, float)):  # Check input
        return str(text)  # Convert to string and return
    stemmer = SnowballStemmer('english')
    return ' '.join([stemmer.stem(word) for word in text.split()])

def lemmatize_text(text):
    if isinstance(text, (int, float)):  # Check if input is an integer or float
        return str(text)  # Convert integer or float to string and return
    lemmatizer = WordNetLemmatizer()
    return ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
df = pd.read_csv('output.csv')
df
```

```python
for column in df.columns:
    df[column] = df[column].apply(remove_special_characters)
df
```

```python
for column in df.columns:
    df[column] = df[column].apply(remove_numbers)
df
```

```python
for column in df.columns:
    df[column] = df[column].apply(remove_stopwords)
df
```

```python
for column in df.columns:
    df[column] = df[column].apply(lowercase_text)
df
```

```python
for column in df.columns:
    df[column] = df[column].apply(stem_text)
df
```

```python
for column in df.columns:
    df[column] = df[column].apply(lemmatize_text)
df
```

# Question 3 (30 points)

Write a python program to **conduct syntax and structure analysis of the clean text** you just saved above. The syntax and structure analysis includes:

(1) **Parts of Speech (POS) Tagging:** Tag Parts of Speech of each word in the text, and calculate the total number of N(oun), V(erb), Adj(ective), Adv(erb), respectively.

(2) **Constituency Parsing and Dependency Parsing:** print out the constituency parsing trees and dependency parsing trees of all the sentences. Using one sentence as an example to explain your understanding about the constituency parsing tree and dependency parsing tree.

(3) **Named Entity Recognition:** Extract all the entities such as person names, organizations, locations, product names, and date from the clean texts, calculate the count of each entity.

```python
df['POS_tags'] = df['Description'].apply(lambda x: nltk.pos_tag(nltk.word_tokenize(x)))

# Calculate Nouns, Verbs, Adjectives, and Adverbs
noun_count = 0
verb_count = 0
adj_count = 0
adv_count = 0

for tags in df['POS_tags']:
    for word, tag in tags:
        if tag.startswith('N'):
            noun_count += 1
        elif tag.startswith('V'):
            verb_count += 1
        elif tag.startswith('J'):
            adj_count += 1
        elif tag.startswith('R'):
            adv_count += 1

print("Total Nouns (N):", noun_count)
print("Total Verbs (V):", verb_count)
print("Total Adjectives (Adj):", adj_count)
print("Total Adverbs (Adv):", adv_count)
```

```python
import spacy
from collections import Counter
import pandas as pd

nlp = spacy.load('en_core_web_sm')

# Function to extract entities and labels from text
def extract_entities(text):
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

# Extract entities from text in the 'Description' column
all_entities = []
for text in df['Description']:
    entities = extract_entities(text)
    all_entities.extend(entities)

# Count the occurrences of each entity type
entity_counts = Counter([entity[1] for entity in all_entities])

# Print the entities and their counts
print("Entities and their counts:")
for entity_type, count in entity_counts.items():
    print(f"{entity_type}: {count}")
```

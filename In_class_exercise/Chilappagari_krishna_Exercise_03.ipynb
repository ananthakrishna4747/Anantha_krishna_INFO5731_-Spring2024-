#POS Feature Extraction

```python
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Input text

# Process the text
doc = nlp(text)

# Part-of-Speech (POS) Tags
pos_tags = [(token.text, token.pos_) for token in doc]

# Print the results
print("Part-of-Speech (POS) Tags:", pos_tags)

```

#Finding Trigrams to get The Context

```python
from nltk import ngrams
import re

tokens = re.findall(r'\b\w+\b', text.lower())

# Generate trigrams
trigrams = list(ngrams(tokens, 3))

print("Trigrams:")
for trigram in trigrams:
    print(trigram)
```

#Finding Relationships between Each word

```python
import spacy

# Load English tokenizer, tagger, parser, NER, and word vectors
nlp = spacy.load("en_core_web_sm")

# Process your document text
doc = nlp(text)

# Extract subject-predicate-object triples using dependency parsing
relationships = []
for token in doc:
    if token.dep_ == "ROOT":
        # Extract the subject and object of the root verb
        subject = [child.text for child in token.children if child.dep_ == "nsubj"]
        object_ = [child.text for child in token.children if child.dep_ == "dobj"]
        if subject and object_:
            relationships.append((subject[0], token.text, object_[0]))

# Print the extracted relationships
for relationship in relationships:
    print(relationship)
```

#Analysing Each word and Context using Spacy

```python
import spacy

nlp = spacy.load("en_core_web_sm")
text = data[0]['body']
# Process the text with spaCy
doc = nlp(text)

# Analyze each word
for token in doc:
    print("\nWord:", token.text)
    print("Lemma:", token.lemma_)

    # Get context words
    context_words = []
    for context_token in token.lefts:
        context_words.append(context_token.text)
    for context_token in token.rights:
        context_words.append(context_token.text)
    print("Context Words:", context_words)
```

#Finding Dependency Tree for the Given document

```python

import spacy
from spacy import displacy
# Plot the dependency graph
doc = nlp(text)
displacy.render(doc, style='dep',jupyter=True)
for tok in doc:
  print(tok.text,"-->",tok.dep_,"-->",tok.pos_)
```

## Question 3 (10 points):
Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)."

Select the most important features you extracted above, rank the features based on their importance in the descending order.

```python
# You code here (Please add comments in the code):

import json
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_selection import chi2

# Load the JSON dataset
with open('/content/output.json', 'r') as json_file:
    data = json.load(json_file)

# Create a DataFrame from the JSON data
df = pd.DataFrame(data)

# Use the "body" column as the feature matrix X
X = df['body']

# Use the "Classification" column as the target variable y
y = df['classification']

# Vectorize the text data
vectorizer = CountVectorizer()
X_vec = vectorizer.fit_transform(X)

# Calculate the Chi-square scores
chi2_scores, p_values = chi2(X_vec, y)

# Create a DataFrame to store the scores
chi2_df = pd.DataFrame({'feature': vectorizer.get_feature_names_out(),
                        'chi2_score': chi2_scores,
                        'p_value': p_values})

# Sort the features by Chi-square score in descending order
chi2_df = chi2_df.sort_values(by='chi2_score', ascending=False)

# Print the top features
print(chi2_df.head(10))

```

## Question 4 (10 points):
Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
# You code here (Please add comments in the code):
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load the BERT model
model = SentenceTransformer('bert-base-nli-mean-tokens')

# Text data
text_data = df['body']

# Query
query = "What legal consequences did the owners of DC Solar, Jeff Carpoff and Paulette Carpoff, face for their involvement in the Ponzi scheme, and what actions did they take to conceal their fraudulent activities?"

# Calculate BERT embeddings for the text data
text_embeddings = model.encode(text_data)

# Calculate BERT embedding for the query
query_embedding = model.encode([query])[0]

# Calculate cosine similarity between the query and each text in the data
similarities = cosine_similarity([query_embedding], text_embeddings)[0]

# Rank the text based on similarity
ranked_indices = np.argsort(similarities)[::-1]

# Print the ranked text
for i, idx in enumerate(ranked_indices):
    print(f"Rank {i+1}: Similarity {similarities[idx]} - {text_data[idx]}")



```
